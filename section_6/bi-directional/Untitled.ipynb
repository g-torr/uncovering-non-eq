```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plus(P):
    return 0.5*(1-np.tanh(theta/np.sqrt(avg_degree)/2/T)+P*(np.tanh((1-theta)/np.sqrt(avg_degree)/2/T)+np.tanh(theta/np.sqrt(avg_degree)/2/T)))
def minus(P):
    return 0.5*(1-np.tanh(theta/np.sqrt(avg_degree)/2/T)+P*(np.tanh((-1-theta)/np.sqrt(avg_degree)/2/T)+np.tanh(theta/np.sqrt(avg_degree)/2/T)))

T = 1.4
avg_degree = 1
theta = 0
P = np.linspace(0,1,100)
plt.plot(P,plus(P))
plt.plot(P,minus(P))
plt.plot([0,1],[0,1],c = 'k')
plt.axhline(0.5,color = 'k')
```

symmetric couplings are positively bias
$$
\rho_+(\rho_+(P))+\rho_-(\rho_-(P))\geq 1
$$
antisymmetric couplings are positively bias
$$
\rho_+(\rho_-(P))+\rho_-(\rho_+(P))\leq 1
$$

```python
P = np.linspace(0,1,100)
plt.plot(P,minus(plus(P)))
plt.plot(P,plus(minus(P)))
plt.plot(P,P,ls = '--',c = 'k')
plt.axhline(0.5,color = 'k')
```
